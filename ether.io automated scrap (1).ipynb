```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import math
```

```python
HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
```

```python
k=1
id=[]
b=[]
t=[]
v=[]
m=[]
txn=[]
from1=[]
while(k<=6):
    url1=f'https://etherscan.io/txs?a=0x1f9090aae28b8a3dceadf281b0f12828e676c326&ps=100&p={k}'
    page=requests.get(url1, headers=HEADERS)
    soup = BeautifulSoup(page.content, "html.parser")
    
    for row in soup.select('table.table-hover tbody tr'):
        cells = row.findAll('td')
        for i in cells[1]:
            id.append(i.text)
        for i in cells[4]:
            m.append(i.text)
        for i in cells[3]:
            b.append(i.text)
        for i in cells[7]:
            from1.append(i.text)
        for i in cells[9]:
            t.append(i.text) 
        for i in cells[10]:
            v.append(i.text) 
        for i in cells[11]:
            txn.append(i.text)
    k=k+1
```

```python
for j in range(len(id)):
    id[j]=id[j].replace('\n','')
    from1[j]=from1[j].replace('\n','')
    t[j]=t[j].replace('\n','')
```

```python
original_list = txn
tax = [''.join(original_list[i:i+3]) for i in range(0, len(original_list), 3)]
```

```python
import pandas as pd 
  
# initialize data of lists. 
data = {'Txn Hash': id, 
        'Block': b,'Time': m,'From': from1,'Value': v,'Txn Fee':tax} 
  
# Create DataFrame 
df = pd.DataFrame(data) 
```
